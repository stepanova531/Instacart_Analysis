# 04 Aggregating data with agg()

```python
# Group the data by department_id and aggregate the data by order_number
df.groupby('department_id').agg({'order_number': ['mean']})
```

# The output shows the average number of orders per user for each department ID.

# 05 Aggregating data with mean()

```python
# Group the data by department_id and aggregate the data by order_number
df.groupby('department_id')['order_number'].mean()
```

```python
df.groupby('department_id').order_number.mean()
```

# 06 Multiple Aggregations

```python
# Group and aggregate using multiple statistics (mean. min, max)
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

# 07 Aggregating Data with transform()

```python
# Create a new column 'max_order' and calculate maximum orders per customer
df_ords_prods_merge['max_order']=df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge['order_number'].loc[df_ords_prods_merge['user_id']==1].describe()
```

```python
# Create a new column 'loyalty_flag' and assign loyalty label 'Loyal customer' if order number >40
df_ords_prods_merge.loc[df_ords_prods_merge['max_order']>40, 'loyalty_flag']='Loyal customer'
```

```python
# Create a new column 'loyalty_flag' and assign loyalty label 'Regular customer' if order nember <=40 and >10 
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order']<=40) & (df_ords_prods_merge['max_order']>10), 'loyalty_flag']='Regular customer'
```

```python
# Create a new column 'loyalty_flag' and assign loyalty label 'New customer' if order nember <=10
df_ords_prods_merge.loc[df_ords_prods_merge['max_order']<=10, 'loyalty_flag']='New customer'
```
