# 03 Using defined function

```python
# Define the function
def price_label(row):
    if row['prices']<=5:
        return 'Low range product'
    elif (row['prices']>5) and (row['prices']<=15):
        return 'Midle range product'
    elif row['prices']>15:
        return 'High range product'
    else: return np.nan
```

```python
# Apply the function
df['price_range']=df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna=False)
```

# 05 Using For-Loops

```python
# Create a loop to range the busiest day of the week
result=[]
for value in df_ords_prods_merge['orders_day_of_week']:
    if value==0:
        result.append("Busiest day")
    elif value==4:
            result.append('Least busy')
    else:
                result.append('Regular busy')
```

# 08 Task 4.7. Step 4. Creating a time labeled column for busiest hours of the day 

```python
# Check for frequency in oreder_hour_of_day column
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna=False)
```

# The first eight values in the frequency list will be labeled as ‘Most orders’ (9, 10, 11, 12, 13, 14, 15, 16) and the last eight hours will be labeled as ‘Fewest orders’ accordingly (0, 1, 2, 3, 4, 5, 6, 23). The eight values in the middle of the list will be added as ‘Average orders’ (7, 8, 17, 18, 19, 20, 21, 22)

```python
# Create a loop to range the busiest and slowest hours of the day
result4=[]
for value in df_ords_prods_merge['order_hour_of_day']:
    if value in [9, 10, 11, 12, 13, 14, 15, 16]:
        result4.append("Most orders")
    elif value in [0, 1, 2, 3, 4, 5, 6, 23]:
            result4.append("Fewest orders")
    else:
                result4.append("Average orders")
```

```python
result4
```

```python
# Create a new column 'busiest_period_of_day' and set it equal to result4
df_ords_prods_merge['busiest_period_of_day']=result4
```

```python
df_ords_prods_merge['busiest_period_of_day'].value_counts(dropna=False)
```
