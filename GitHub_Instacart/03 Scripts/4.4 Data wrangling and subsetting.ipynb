```python
# Breakfast time sales in Orders 
df_ords_bf=df_ords.loc[df_ords['order_hour_of_day'].isin([6,7,8,9])]
```

```python
df_ords_bf.describe()
```

```python
# Breakfast time sales in Orders on weekend supposing 6 assigned to Saturday and 0 assignd to Sunday 
df_ords_bf_wend=df_ords.loc[(df_ords['orders_day_of_week'].isin([0,6]))&(df_ords['order_hour_of_day'].isin([6,7,8,9]))]
```

```python
df_ords_bf_wend.describe()
```

```python
# Breakfast time sales in Orders on workdays supposing 1,2,3,4,5 assigned to workdays from Monday to Friday 
df_ords_bf_workd=df_ords.loc[(df_ords['orders_day_of_week'].isin([1,2,3,4,5]))&(df_ords['order_hour_of_day'].isin([6,7,8,9]))]
```

```python
df_ords_bf_workd.describe()
```

```python
# Count of values in 'day of week' column
df_ords['orders_day_of_week'].value_counts(dropna=False)
```

# The subsets above show the number of orders placed during breakfast, presumably between 6am and 9am during the week and separately on weekends and weekdays. The number of orders on Saturday and Sunday is 27% of the total number of orders. I expected to see most orders placed on weekends. However, the frequency is likely distributed evenly throughout the week.
