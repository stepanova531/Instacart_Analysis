# Step 4. Creating a loyalty flag for customers

```python
# Create a new column 'max_order' and calculate maximum orders per customer
df_ords_prods_merge['max_order']=df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df_ords_prods_merge[['user_id','max_order']].head()
```

```python
# Create a new column 'loyalty_flag' and assign loyalty label 'Loyal customer' if order number >40
df_ords_prods_merge.loc[df_ords_prods_merge['max_order']>40, 'loyalty_flag']='Loyal customer'
```

```python
# Create a new column 'loyalty_flag' and assign loyalty label 'Regular customer' if order nember <=40 and >10 
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order']<=40) & (df_ords_prods_merge['max_order']>10), 'loyalty_flag']='Regular customer'
```

```python
# Create a new column 'loyalty_flag' and assign loyalty label 'New customer' if order nember <=10
df_ords_prods_merge.loc[df_ords_prods_merge['max_order']<=10, 'loyalty_flag']='New customer'
```

```python
df_ords_prods_merge.shape
```

```python
df_ords_prods_merge.head()
```

```python
# Make a correction: remove a column 'average_spend' previously created on a wrong step
df_ords_prods_merge=df_ords_prods_merge.drop(columns=['average_spend'])
```

# Step 6. Creating a Spending flag for customers

```python
# Create a new column 'spender_flag' and assign the label 'High spender' if 'average_spending' >=10
df_ords_prods_merge.loc[df_ords_prods_merge['average_spending']>=10, 'spender_flag']='High spender'
```

# Step 7. Creating an order frequency flag for customers  

```python
# Check for the values in 'days_since_prior_order' column
df_ords_prods_merge['days_since_prior_order'].describe()
```

```python
df_ords_prods_merge['days_since_prior_order'].isnull().sum()
```

# There are 2,075,745 NaNs values related to the first order placed by the client.

```python
# Create a new column 'median_days_prior_order' and calculate the median of 'days_since_prior_order' per customer
df_ords_prods_merge['median_days_prior_order']=df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check for the dataframe after adding a new column 'median_days_prior_order'
df_ords_prods_merge.shape
```

```python
df_ords_prods_merge.head()
```

```python
# Create a new column 'order_frequency_flag' and assign the label 'Non-frequent customer' if the median >20
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_prior_order']>20, 'order_frequency_flag']='Non-frequent customer'
```

```python
# Create a new column 'order_frequency_flag' and assign the label 'Regular customer' if the median >10 and <=20
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_prior_order']>10) & (df_ords_prods_merge['median_days_prior_order']<=20), 'order_frequency_flag']='Regular customer'
```

```python
# Create a new column 'order_frequency_flag' and assign the label 'Frequent customer' if the median <=10
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_prior_order']<=10, 'order_frequency_flag']='Frequent customer'
```

```python
# Check for the dataframe after adding a new column 'order_frequency_flag'
df_ords_prods_merge.shape
```

```python
df_ords_prods_merge.head()
```

```python
# Count the frequency of values on Order frequency flag column
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna=False)
```
